```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline 

# key = number of samples, value = pipeline time in minutes
sample_times = {300: 16, 500000: 93, 750000: 171, 1000000: 276}
x = np.array(list(sample_times.keys()))
y = np.array(list(sample_times.values()))

x_new = np.linspace(x.min(), x.max(), 300) 
gfg = make_interp_spline(x, y, k=3) 
ynew = gfg(x_new)
plt.scatter(sample_times.keys(), sample_times.values(), color='red',zorder=2)
plt.plot(x_new, ynew, zorder=1)

plt.title('Simulation time')
plt.xlabel('Number of samples (million)')
plt.ylabel('Time taken (minutes)')
plt.show()
```
