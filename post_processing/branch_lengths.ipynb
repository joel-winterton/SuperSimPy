```python
import matplotlib.pyplot as plt
from Bio import Phylo
import numpy as np
```

```python
sim_tree = Phylo.read('/home/joel/Downloads/sim_1mil_v2.tree','newick')
```

```python
covid_tree = Phylo.read('/home/joel/Downloads/Viridian_2M_SARS-CoV-2_MAPLE_tree.tree/Viridian_2M_noShortDel_MAPLE_tree.tree', 'newick')
```

```python
def tip_distances(root):
    stack = [(root, 0)]
    distances = []

    while stack:
        node, current_distance = stack.pop()

        if node.is_terminal():
            distances.append(current_distance)
        else:
            for clade in node.clades:
                stack.append((clade, current_distance + int(clade.branch_length)))

    return distances
```

```python
sim_distances = tip_distances(sim_tree.clade)
```

```python
covid_distances = tip_distances(covid_tree.clade)
```

```python
plt.hist(sim_distances, bins=10)
plt.show()
```
